```python
from sklearn.linear_model import LogisticRegressionCV
from sklearn.preprocessing import StandardScaler

import pandas as pd
import numpy as np
```

```python
labels = pd.read_csv("../data/human/CRC_metadata.csv")

file_path = "../data/Priya_et_al_Supplementary_Tables/Supplementary Tables S12-S17.xlsx"

data = pd.read_excel(file_path, sheet_name=None)
data.keys()
```

```python
gene_data = data["S12"]
gene_data.head()
```

```python
gene_data = gene_data.set_index('Unnamed: 0')
gene_data.head()
```

T-Test:

```python
from scipy.stats import ttest_ind
```

```python
y = labels["Description"].map({"tumor": 1, "normal": 0}).to_numpy()
y
```

```python
x = gene_data.to_numpy()
x = x[:,:]
x
```

```python
scaler = StandardScaler()
x = scaler.fit_transform(x)
x = pd.DataFrame(x, columns=gene_data.columns, index=gene_data.index)
```

```python


p_values = []
for gene in gene_data.columns:
    t_stat, p = ttest_ind(x[gene], x[gene], equal_var=False)
    p_values.append(p)

# Add p-values to a DataFrame
ttest_results = pd.DataFrame({
    "gene": x.columns,
    "p_value": p_values
}).set_index("gene")
```

```python
ttest_results
```

```python
# Fit LASSO with cross-validation
lasso = LogisticRegressionCV(
    penalty='l1', solver='saga', cv=5, max_iter=10000, random_state=42
)
lasso.fit(x.to_numpy().T, y)


```

```python
lasso.coef_[0]
```

```python
# Get coefficients
coef = pd.Series(lasso.coef_[0], gene_data.index)
selected_genes = coef[coef != 0]
```

```python
selected_genes
```

```python
selected_genes_sorted = selected_genes.sort_values(ascending=False)

selected_genes_sorted
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    x.to_numpy().T, y,
    test_size=0.2,        # 20% test data, 80% train
    random_state=42,      # ensures reproducibility
    stratify=y            # keeps class distribution balanced
)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
lasso_predictor = LogisticRegressionCV(
    penalty='l1', solver='saga', cv=5, max_iter=10000, random_state=42
)
lasso_predictor.fit(X_train, y_train)
```

```python
y_pred = lasso_predictor.predict(X_test)
y_proba = lasso_predictor.predict_proba(X_test)[:, 1]
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test, y_pred))
```
